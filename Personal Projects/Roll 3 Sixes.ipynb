# Roll 3 Sixes

I though it would be interesting to use some of the data science methodology I've been learning to do an experiment.
How often will 3 sixes come up on a set of 3 dice.

The odds are easy to calculate. 1/6 * 1/6 * 1/6 = 1/216

There are 216 total combinations that can be rolled with 3 dice. Only 1 of those is 3 sixes.

So we are going to use a random number generator to generate the roll of 3 dice. We're going to roll the 3 dice 216 times. We're going to track home many times 3 sixes comes up. We're also going to track when the first 3 sixes comes up.

Of course performing the experiment once won't really give us a clear answer, so we'll increase our sample size to get more data points. We'll perform our test in increments of: 1, 10, 100, 1000, 10,000 and compare our results



```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
test = 1 # The number of times we'll be rolling the sequence of 216 rolls

for loop in range(test):
    first_sixes = []    # A list of which roll resulted in the first 3 sixes
    all_sixes = []      # A list of each roll that resulted in 3 sixes
    sixes_counter = []
    counter = 0   # The number of times 3 sixes was rolled
    
    for roll in range(216):
        a = random.randint(1,6)
        b = random.randint(1,6)
        c = random.randint(1,6)
        
        if a+b+c == 18:
            all_sixes.append(roll+1)
            counter += 1
            if counter == 1:
                first_sixes.append(roll+1)
    sixes_counter.append(counter)
        
print('The first set of 3 sixes was on roll:', first_sixes[:])
print('A roll of 3 sixes was rolled', sixes_counter[:], 'times')
print('A roll of 3 sixes appeared on the following rolls:', all_sixes[:])
            
    
```

```python
test = 10 # The number of times we'll be rolling the sequence of 216 rolls

first_sixes = []    # A list of which roll resulted in the first 3 sixes
all_sixes = []      # A list of each roll that resulted in 3 sixes
sixes_counter = []  # How many times 3 sixes were rolled out of 216
    
for loop in range(test):
    
    counter = 0   # The number of times 3 sixes was rolled
    
    for roll in range(216):
        a = random.randint(1,6)
        b = random.randint(1,6)
        c = random.randint(1,6)
        if a+b+c == 18:
            all_sixes.append(roll+1)
            counter += 1
            if counter == 1:
                first_sixes.append(roll+1)
    sixes_counter.append(counter)
        

print('The first set of 3 sixes was on roll:', first_sixes[:])
print('A roll of 3 sixes was rolled', sixes_counter[:], 'times')
print('A roll of 3 sixes appeared on the following rolls:', all_sixes[:])

print()
x = np.arange(217)
plt.figure(figsize=(20,10))
plt.title('The First Roll of 3 sixes', fontsize=36)
plt.ylabel('Number of Occurrences', fontsize=36)
plt.xlabel('The Roll Out of 216', fontsize=36)
plt.xlim(0, 216)
plt.grid(axis='y')
plt.hist(first_sixes)
plt.show()
print()

plt.figure(figsize=(20,10))
plt.title('How Many Times 3 Sixes Were Rolled', fontsize=36)
plt.ylabel('Number of Occurrences', fontsize=36)
plt.xlabel('Ten Games of 216 Rolls', fontsize=36)
plt.xlim(-1, 10)
plt.grid(axis='y')
plt.hist(sixes_counter, bins=12)
plt.show()
print()

plt.figure(figsize=(20,10))
plt.title('The Rolls on which 3 Sixes Occured', fontsize=36)
plt.ylabel('Number of Occurrences', fontsize=36)
plt.xlim(-1, 216)
plt.grid(axis='y')
plt.hist(all_sixes, bins=218)
plt.show()
```
